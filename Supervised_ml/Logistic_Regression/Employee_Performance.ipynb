```python
y_bar=np.array([df[df['Attrition']=='No'].shape[0],df[df['Attrition']=='Yes'].shape[0]])
x_bar=['No (0)','Yes (1)']
plt.bar(x_bar,y_bar)
plt.xlabel('Labels/Classes')
plt.ylabel('Number of Instances')
plt.title('Distribution of Labels classes in the Datset') 
df.drop(['EmpNumber','Age'],axis='columns',inplace=True)
```

```python
df['Attrition'] = df['Attrition'].map({'Yes':1,'No':0})
y=df.iloc[:,1].values
```

```python
x=pd.get_dummies(data=df,columns=['BusinessTravelFrequency','EmpDepartment','EducationBackground','Gender','EmpJobRole','EmpLastSalaryHikePercent',
                                 'EmpJobSatisfaction','TotalWorkExperienceInYears','ExperienceYearsAtThisCompany','ExperienceYearsInCurrentRole',
                                 'YearsSinceLastPromotion','YearsWithCurrManager','MaritalStatus','Attrition'],drop_first=True)
```

```python
df
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42,shuffle=True)
regressor=LogisticRegression()
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
print(regressor.score(x_test,y_test))
print(regressor.score(x_train,y_train))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
confusion_matrix=confusion_matrix(y_test,y_pred)
confusion_matrix
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
