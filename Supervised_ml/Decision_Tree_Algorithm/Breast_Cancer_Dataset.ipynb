```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_breast_cancer
```

```python
dataset=load_breast_cancer()
dataset
```

```python
df=pd.DataFrame(dataset['data'],columns=dataset['feature_names'])
df
```

```python
df['Cancer_Type']=dataset['target']
df
```

```python
x=df.drop('Cancer_Type',axis='columns')
y=dataset['target']
```

```python
print(f"shape of x:{x.shape}")
print(f"shape of y:{y.shape}")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
print(f'x_train shape{x_train.shape}')
print(f'x_test shape{x_test.shape}')
print(f'y_train shape{y_train.shape}')
print(f'y_test shape{y_test.shape}')                   
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
imp=model.feature_importances_
print(imp)
name=dataset['feature_names']
name
```

```python
plt.figure(figsize=(15,10))
plt.barh(name,imp)
```

```python
df_new=pd.DataFrame(df[['worst concave points','worst perimeter','worst texture','worst radius','fractal dimension error','perimeter error','texture error','radius error','mean concave points','mean smoothness','mean texture']])
df_new.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain1,xtest1,ytrain1,ytest1= train_test_split(df_new.loc[:,:],y,test_size=0.30,random_state=1)
print(f'xtrain shape{xtrain1.shape}')
print(f'xtest shape{xtest1.shape}')
print(f'ytrain shape{ytrain1.shape}')
print(f'ytest shape{ytest1.shape}')
```

```python
n_model=DecisionTreeClassifier()
n_model.fit(xtrain1,ytrain1)
```

```python
pv=model.predict(x_test)
pv1=n_model.predict(xtest1)
```

```python
pv1
```

```python
from sklearn.metrics import accuracy_score
```

```python
print(f'accuracy of model1 :{accuracy_score(y_test,pv)}')
print(f'accuracy of model2 :{accuracy_score(ytest1,pv1)}')
```
